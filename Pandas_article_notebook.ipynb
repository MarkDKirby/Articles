# Interpoation of Charging Outlets

## Input Some Data

```python
c = [21,125,627,4235,17843,35892,56934,74351,92821,110631,128556]
```

```python
outlets = []
for i in range(0,((len(c)-1)*12)+1):
    if i%12==0:
        outlets.append(c[int(i/12)])
    else:
        outlets.append(np.nan)
outlets
```

## Make a DataFrame

```python
charging_outlets = pd.DataFrame(outlets, columns = ["Charging Outlets"])
charging_outlets
```

## Linear Interpolation

```python
charging_outlets["linear"] = charging_outlets.interpolate(method="linear", inplace=False)

charging_outlets
```

```python
charging_outlets["l_rounded"] = np.rint(charging_outlets["linear"])
charging_outlets['l_rounded'] = charging_outlets["l_rounded"].astype(int)
```

```python
charging_outlets
```

## Quadratic Interpoation

```python
charging_outlets['quadratic'] = charging_outlets["Charging Outlets"].interpolate(method="quadratic", inplace=False)
```

```python
charging_outlets["q_rounded"] = np.round(charging_outlets["quadratic"])
charging_outlets["q_rounded"] = charging_outlets["q_rounded"].astype(int)
```

```python
charging_outlets
```

## Spline Interpolation

```python
charging_outlets['spline3'] = charging_outlets["Charging Outlets"].interpolate(method="spline", order=3, inplace=False)
charging_outlets["s3_rounded"] = np.round(charging_outlets["spline3"])
charging_outlets["s3_rounded"] = charging_outlets["s3_rounded"].astype(int)
charging_outlets
```

## Polinomial Interpolation

```python
charging_outlets['polynomial5'] = charging_outlets["Charging Outlets"].interpolate(method="polynomial", order=5, inplace=False)
charging_outlets["p5_rounded"] = np.round(charging_outlets["polynomial5"])
charging_outlets["p5_rounded"] = charging_outlets["p5_rounded"].astype(int)
charging_outlets
```

```python
month = []
for i in range(0,len(charging_outlets)):
    month.append(i)
print(month[0],month[-1])
```

```python
charging_outlets["month"] = month
charging_outlets
```

## Plotting the Data

```python
plt.plot(charging_outlets["month"],charging_outlets["l_rounded"], label = "Linear")
plt.plot(charging_outlets["month"],charging_outlets["q_rounded"], label = "Quadratic")
plt.plot(charging_outlets["month"],charging_outlets["s3_rounded"], label = "Spline")
plt.plot(charging_outlets["month"],charging_outlets["p5_rounded"], label = "Polynomial")
plt.scatter(charging_outlets["month"],charging_outlets["Charging Outlets"], label = "Yearly Data")
plt.title("Nan_Value Fills by Type")
plt.xlabel("Month")
plt.ylabel("Charging Outlets")
plt.legend()
plt.rc('font', size=10) 
plt.rc('axes', titlesize=30) 
plt.rc('axes', labelsize=30) 
plt.rc('xtick', labelsize=30) 
plt.rc('ytick', labelsize=30) 
plt.rc('legend', fontsize=30) 
plt.rcParams["figure.figsize"] = (20,20)
plt.show()
```
